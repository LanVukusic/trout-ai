```python
import os

DATA_URL = "https://database.lichess.org/standard/lichess_db_standard_rated_2017-09.pgn.zst"
OUT_FILE = "lichess_data.pgn.zst"

# download if not exists
if not os.path.exists(OUT_FILE):
    !wget -O $OUT_FILE $DATA_URL

# unzip
if not os.path.exists(OUT_FILE[:-4]):
    !zstd -d $OUT_FILE



```

```python
# use python chess library to parse the pgn file

import chess.pgn
import random

# import local descriptions lib
from descriptions import generate_captures_texts
from descriptions import generate_checks_texts
from descriptions import generate_pinned_texts

SAMPLE_CHANCE = 1

with open(OUT_FILE[:-4]) as pgn:
    game = chess.pgn.read_game(pgn)
    while game:
        # print(game.headers["White"], game.headers["Black"], game.headers["Result"])
        boards = []

        board = game.board()
        for i, move in enumerate(game.mainline_moves()):
          # input chess move to board
          board.push(move)
          if random.random() < SAMPLE_CHANCE:
            boards.append(board.copy())

        for board in boards:
          # input chess move to board
          # print(generate_captures_texts(board, include_pawns=False))
          # print(generate_checks_texts(board))
          # print(generate_pinned_texts(board))

          descriptions = []
          descriptions.extend(generate_captures_texts(board, include_pawns=False))
          descriptions.extend(generate_checks_texts(board))
          descriptions.extend(generate_pinned_texts(board))

          # pick 5 random descriptions
          random.shuffle(descriptions)
          descriptions = descriptions[:5]

          print(descriptions)
        
        # read next game
        game = chess.pgn.read_game(pgn)

        if(input() == "q"):
          break

```
